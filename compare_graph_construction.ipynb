```python
import numpy as np
```

# Load performance metrics

```python
# Graph built on subject 4 (original)
avg_acc_test_s4 =np.load("./graph_construction_metrics/avg_acc_test_s4.npy")
avg_acc_train_s4 = np.load("./graph_construction_metrics/avg_acc_train_s4.npy")
avg_loss_test_s4 = np.load("./graph_construction_metrics/avg_loss_test_s4.npy")
```

```python
# Graph built on subject 2
avg_acc_test_s2 =np.load("./graph_construction_metrics/avg_acc_test_s2.npy")
avg_acc_train_s2 = np.load("./graph_construction_metrics/avg_acc_train_s2.npy")
avg_loss_test_s2 = np.load("./graph_construction_metrics/avg_loss_test_s2.npy")
```

```python
# Graph built on subject 2 (z-transformed)
avg_acc_test_s2_z =np.load("./graph_construction_metrics/avg_acc_test_s2_z.npy")
avg_acc_train_s2_z = np.load("./graph_construction_metrics/avg_acc_train_s2_z.npy")
avg_loss_test_s2_z = np.load("./graph_construction_metrics/avg_loss_test_s2_z.npy")
```

# Plot performance metrics for different graph construction methods

```python
import matplotlib.pyplot as plt

epochs = 70

fig = plt.figure(figsize=(7,6))
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_train_s4, c = "blue", label = "S4, Train", ls = 'dotted')
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_test_s4, c = "red", label = "S4, Test", ls = 'dotted')
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_loss_test_s4, c = "green", label = "S4, Loss", ls = 'dotted')

plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_train_s2, c = "blue", label = "S2, Train", ls = "dashed")
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_test_s2, c = "red", label = "S2, Test", ls = "dashed")
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_loss_test_s2, c = "green", label = "S2, Loss", ls = "dashed")

plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_train_s2_z, c = "blue", label = "S2, z-transformed, Train")
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_acc_test_s2_z, c = "red", label = "S2, z-transformed, Test")
plt.plot(np.asarray(np.arange(epochs) + 1).astype("str"), avg_loss_test_s2_z, c = "green", label = "S2, z-transformed, Loss")
plt.ylabel("Acc/Loss")
plt.xlabel("Epoch")

plt.xticks(np.asarray(np.arange(epochs) + 1).astype("str")[::3])
fig.legend(bbox_to_anchor=(0, 0, 0.9, 0.6))
```

Performance is best when the graph is constructed from subject 2 and the connectivity matrix is z-transformed.
